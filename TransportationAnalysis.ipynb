```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
file = "./data/port_traffic.csv"

commercialTraffic_file = './data/commercial_traffic_by_industry.csv'
```

```python
port_traffic = pd.read_csv(file)

commerical_traffic = pd.read_csv(commercialTraffic_file)
```

```python
port_traffic
```

```python
#Get the change in trip volume (how much is being shipped) by date

fig, ax = plt.subplots(4, 1, figsize=(20,10))
fig.tight_layout()

ports = [1, 2, 3, 4]


for i in range(0, len(ports)):

    port_traffic_perport = port_traffic.loc[port_traffic['port_id'] == (i+1)]

    port_traffic_resorted = port_traffic_perport.sort_values(by=['week_end'])

    ax[i].plot(port_traffic_resorted['week_end'], port_traffic_resorted['percent_of_trip_volume_change'], '--', color = 'r')
    #ax[i].plot(port_traffic_resorted['week_end'], port_traffic_resorted['wait_time_hdt'], '--', color = 'b')
    #ax[i].plot(port_traffic_resorted['week_end'], port_traffic_resorted['wait_time_mdt'], '--', color = 'g')
    #ax[i].plot(port_traffic_resorted['week_end'], port_traffic_resorted['wait_time_ldt'], '--', color = 'm')
        
    
    if i != 3:
        ax[i].tick_params(labelbottom = False)
    else:
        ax[-1].set_xticklabels(labels = port_traffic_resorted['week_end'] , rotation='vertical');
        
    ax[i].axvline(x = "2020-03-07", color = 'b')
        
    
    ax[i].set_title("Port "+ str(ports[i]))
    
    
plt.savefig("./figs/percent_change_in_trip_volume_NorthAmerica.png")
```

```python
#Get percentage changes for (hdt, mdt, and ldt)

#Get the change in trip volume (how much is being shipped) by date

fig, ax = plt.subplots(4, 1, figsize=(20,10))
fig.tight_layout()

ports = [1, 2, 3, 4]


for i in range(0, len(ports)):

    port_traffic_perport = port_traffic.loc[port_traffic['port_id'] == (i+1)]

    port_traffic_resorted = port_traffic_perport.sort_values(by=['week_end'])

    ax[i].plot(port_traffic_resorted['week_end'], port_traffic_resorted['percent_of_ldt_volume_change'], '--', color = 'b')
    ax[i].plot(port_traffic_resorted['week_end'], port_traffic_resorted['percent_of_mdt_volume_change'], '--', color = 'g')
    ax[i].plot(port_traffic_resorted['week_end'], port_traffic_resorted['percent_of_hdt_volume_change'], '--', color = 'm')
        
    
    if i != 3:
        ax[i].tick_params(labelbottom = False)
    else:
        ax[-1].set_xticklabels(labels = port_traffic_resorted['week_end'] , rotation='vertical');
        
    ax[i].axvline(x = "2020-03-07", color = 'b')
        
    
    ax[i].set_title("Port "+ str(ports[i]))
    
    
    
plt.savefig("./figs/percent_change_in_trip_volume_NorthAmerica_differentvehicles.png")
```

```python
#Get percent change from Baseline for North America and Industry: Freight & Logistics Services

dateUpTo = '2020-08-01'
    
fig, ax = plt.subplots(2, 1, figsize=(20,10))
fig.tight_layout()

# ======= Plot "Freight & Logistics Services" in "North America"
commerical_traffic_NA_freight = commerical_traffic.loc[(commerical_traffic['region'] == "North America") &
                                                      (commerical_traffic['industry'] == "Freight & Logistics Services") & 
                                                      (commerical_traffic['date'] < dateUpTo)].sort_values(by=['date'])






ax[0].plot(commerical_traffic_NA_freight['date'], commerical_traffic_NA_freight['percent_of_baseline'], marker = 'o', color = 'k')
ax[0].tick_params(labelbottom = False)
#ax[0].set_xticklabels(labels = commerical_traffic_NA_freight['date'] , rotation='vertical');
ax[0].axvline(x = "2020-03-01", color = 'b')
ax[0].set_title("Percent change in baseline Freight & Logistics Services: " +dateUpTo)



# ======= Plot "Transportation (non-freight)" in "North America"
commerical_traffic_NA_nonfreight = commerical_traffic.loc[(commerical_traffic['region'] == "North America") &
                                                      (commerical_traffic['industry'] == "Transportation (non-freight)") & 
                                                      (commerical_traffic['date'] < dateUpTo)].sort_values(by=['date'])



ax[1].plot(commerical_traffic_NA_nonfreight['date'], commerical_traffic_NA_nonfreight['percent_of_baseline'], marker = 'o', color = 'k')
ax[1].set_xticklabels(labels = commerical_traffic_NA_nonfreight['date'] , rotation='vertical');
ax[1].axvline(x = "2020-03-07", color = 'b')
ax[1].set_title("Percent change from Baseline in Transportation (non-freight): " + dateUpTo)


plt.savefig("./figs/percentChangeFromBaseline_NA.png")
```
